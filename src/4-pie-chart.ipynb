### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

def plot_pie_chart():
    try:
        labels = ['Python', 'Java', 'C++', 'JavaScript']
        sizes = [40, 25, 20, 15]
        colors = ['blue', 'green', 'red', 'purple']
        explode = (0.1, 0, 0, 0)  # Highlight the first slice

        if sum(sizes) != 100:
            raise ValueError("Sizes should sum to 100%.")

        plt.figure(figsize=(6, 6))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, shadow=True)
        plt.title("Programming Language Popularity")
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

plot_pie_chart()
```
