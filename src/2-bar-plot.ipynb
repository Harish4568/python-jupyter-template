### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

def plot_bar_chart():
    try:
        categories = ['A', 'B', 'C', 'D']
        values = [5, 7, 3, 9]
        colors = ['red', 'blue', 'green', 'yellow']
        
        if len(categories) != len(values):
            raise ValueError("Mismatch in categories and values length.")

        plt.figure(figsize=(6, 4))
        bars = plt.bar(categories, values, color=colors)

        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom', fontsize=10)

        plt.xlabel('Categories')
        plt.ylabel('Values')
        plt.title('Bar Plot with Labels')
        plt.show()
    
    except Exception as e:
        print(f"Error: {e}")

plot_bar_chart()
```
