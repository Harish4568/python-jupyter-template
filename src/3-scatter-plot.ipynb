### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

def plot_scatter():
    try:
        x = [1, 2, 3, 4, 5]
        y = [2, 4, 6, 8, 10]

        if len(x) != len(y):
            raise ValueError("x and y lists must be of the same length.")

        plt.figure(figsize=(6, 4))
        plt.scatter(x, y, color='blue', marker='o', s=80, edgecolors='black', label="Data points")

        # Adding annotations
        for i in range(len(x)):
            plt.text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=10, ha='right', va='bottom')

        plt.grid(True, linestyle='--', linewidth=0.5)
        plt.xlabel('x-axis')
        plt.ylabel('y-axis')
        plt.title('Scatter Plot with Annotations')
        plt.legend()
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

plot_scatter()
```
