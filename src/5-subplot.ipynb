### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

def plot_subplots():
    try:
        x = [1, 2, 3, 4, 5]
        y = [2, 4, 6, 8, 10]
        categories = ['A', 'B', 'C', 'D']
        values = [5, 7, 3, 9]
        scatter_x = [1, 2, 3, 4, 5]
        scatter_y = [3, 6, 2, 8, 7]
        labels = ['Python', 'Java', 'C++', 'JavaScript']
        sizes = [40, 25, 20, 15]

        fig, axes = plt.subplots(2, 2, figsize=(10, 8))

        # Line plot
        axes[0, 0].plot(x, y, marker='o', linestyle='-', color='blue')
        axes[0, 0].set_title("Line Chart")

        # Bar chart
        axes[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'purple'])
        axes[0, 1].set_title("Bar Chart")

        # Scatter plot
        axes[1, 0].scatter(scatter_x, scatter_y, color='orange', marker='o')
        axes[1, 0].set_title("Scatter Plot")

        # Pie chart
        axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'purple'])
        axes[1, 1].set_title("Pie Chart")

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

plot_subplots()
```
