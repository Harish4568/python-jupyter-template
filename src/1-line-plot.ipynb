### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt

def plot_squares():
    try:
        x = list(range(11))
        y = [i ** 2 for i in x]
        
        if len(x) != len(y):
            raise ValueError("Length of x and y must be the same.")

        plt.figure(figsize=(6, 4))
        plt.plot(x, y, marker='o', linestyle='-', color='b', linewidth=2, markersize=6, label="y = x²")
        plt.grid(True, linestyle='--', linewidth=0.5)
        plt.xlabel('x-axis')
        plt.ylabel('y-axis')
        plt.title('Squares of x')
        plt.legend()
        plt.show()
    
    except Exception as e:
        print(f"Error: {e}")

plot_squares()
```
